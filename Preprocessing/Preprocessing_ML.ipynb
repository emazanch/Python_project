```python
prod
```

```python
prod = prod.rename(columns={'Mois':'Date'})
```

```python
prod['Date'] = pd.to_datetime(prod['Date'], utc=True)
meteo['Date'] = pd.to_datetime(meteo['Date'], utc=True)
```
